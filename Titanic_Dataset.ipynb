# **Анализ датасета титаника**

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

## Утерянные данные

```python
titanic[titanic['Age'].isna()]
```

Данные пассажиров

```python
a = np.sum(titanic['Age'].isna())
print(f'{a} - данных о возрасте потеряно')
```

Данные кают

```python
b = np.sum(titanic['Cabin'].isna())
print(f'{b} - данных о каютах потеряно')
```

Данные погрузки

```python
d = np.sum(titanic['Embarked'].isna())
print(f'{d} - данных о погрузке утеряно')
```

Пассажиры без данных о возрасте и кают

```python
c = np.sum(titanic['Age'].isna() & titanic['Cabin'].isna())
print(f'{c} - данных утеряно')
```

```python
titanic.isnull().sum()
```

## Выжевшие

```python
titanic[titanic['Survived'] == 1]
```

```python
passengers = np.count_nonzero(titanic['PassengerId'])
kid = np.count_nonzero(titanic['Age'] <= 15)
survived = np.count_nonzero(titanic['Survived'])
percent_survived = (survived / passengers) * 100
p_survived = round (percent_survived, 2)
```

Данные о выживших

```python
print(f'Всего пассажиров: {passengers}')
print(f'Среди них дети: {kid}')
print(f'Всего выжило: {survived}')
print(f'Процент выживших: {p_survived}')
```

## Цены на билеты

```python
price = titanic.sort_values (by = ['Fare'])
price
```

Данные о ценах



```python
prices = titanic['Fare']
print(f'Мин. цена: {prices.min()}')
print(f'Средняя. цена: {prices.median()}')
print(f'Макс. цена: {prices.max()}')
```
